```python
users=pd.read_csv('BX-Users.csv', sep=";",error_bad_lines=False, encoding='latin-1')
```

```python
books=pd.read_csv('BX-Books.csv', sep=";",error_bad_lines=False, encoding='latin-1')
```

```python
ratings=pd.read_csv('BX-Book-Ratings.csv', sep=";",error_bad_lines=False, encoding='latin-1')
```

```python
books.drop(['Image-URL-S','Image-URL-M','Image-URL-L'],axis=1,inplace=True)
```

```python
books.rename(columns={'Book-Title':'title','Book-Author':'author','Year-Of-Publication':'year','Publisher':'publisher'},inplace=True)
```

```python
ratings.rename(columns={'User-ID':'user_id','Book-Rating':'rating'},inplace=True)
```

```python
users.rename(columns={'User-ID':'user_id','Location':'location','Age':'age'},inplace=True)
```

```python
a=ratings['user_id'].value_counts()
a
```

```python
b=ratings['user_id'].value_counts()>200
b
```

```python
a[b]
```

```python
new_users=list(a[b].index)
```

```python
len(new_users)
```

```python
ratings=ratings[ratings['user_id'].isin(new_users)]
ratings
```

```python
ratings.shape
```

```python
ratings_with_bookname=ratings.merge(books,on='ISBN')
```

```python
ratings_with_bookname.shape
```

```python
ratings_with_bookname.head()
```

```python
book_num_ratings=ratings_with_bookname.groupby('title')['rating'].count().reset_index()
book_num_ratings
```

```python
book_num_ratings.rename(columns={'rating':'num_ratings'},inplace=True)
```

```python
book_num_ratings=book_num_ratings[book_num_ratings['num_ratings']>=50]
book_num_ratings
```

```python
final_rating=ratings_with_bookname.merge(book_num_ratings,on='title')
final_rating
```

```python
final_rating.shape
```

```python
final_rating=final_rating.drop_duplicates(subset=['user_id','title'])
final_rating
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='rating')
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot=book_pivot.fillna(0)
```

```python
book_pivot
```

```python
#as there are too many zeros in the pivot table 
# so that we convert the pivot table into sparse matrix which will reduce the no. of calculation to be done
from scipy.sparse import csr_matrix
book_matrix=csr_matrix(book_pivot)
book_matrix
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_matrix)
```

```python
book_pivot.index[237]
```

```python
book_pivot.iloc[237,:].values.reshape(1,-1)
```

```python
distance,suggestion=model.kneighbors(book_pivot.iloc[1,:].values.reshape(1,-1),n_neighbors=5)
```

```python
suggestion
```

```python
book_pivot.index[184]
```

```python
for i in suggestion[0]:
    print(book_pivot.index[i])
```

```python
np.where(book_pivot.index=='Exclusive')[0]
```

```python
def recommend(bookname):
    b_id=np.where(book_pivot.index==bookname)[0][0]
    distance,suggestion=model.kneighbors(book_pivot.iloc[b_id,:].values.reshape(1,-1),n_neighbors=5)
    for i in suggestion[0]:
        print(book_pivot.index[i])
```

```python
recommend('The Cradle Will Fall')
```
